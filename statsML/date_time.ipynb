```python
#Working with date table
#Converting to dattime datatype

date['date'] = pd.to_datetime(date['date'])
date.info()
```

```python
#Part 1 - Extract year
date['year'] = date['date'].dt.year
date.head()
```

```python
#Part 2 - Extract month

date['month'] = date['date'].dt.month
date.head()
```

```python
date['month_name'] = date['date'].dt.month_name()
date.head()
```

```python
#Part 3 - Extract date
date['date_day'] = date['date'].dt.day
date.head()
```

```python
date['date_dow'] = date['date'].dt.day_of_week
date.head()
```

```python
date['day_name'] = date['date'].dt.day_name()
date.head()
```

```python
date.columns
```

```python
date.drop(columns=['product_id','city_id','orders'], inplace=True)
date.head()
```

```python
import numpy as np
```

```python
#Weekend means 1 else 0
date['is_weekend'] = np.where(date['day_name'].isin(['Sunday','Saturday']),1,0)
date.head()
```

```python
#Week of the year
date['date_week'] = date['date'].dt.week
date.head()
```

```python
date['quarter'] = date['date'].dt.quarter
date.head()
```

```python
date['half_yearly'] = np.where(date['quarter'].isin([1,2]),1,2)
date.head()
```
