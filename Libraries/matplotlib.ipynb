# Matplotlib

- Mathplotlib is a python library that is used to plot the data
- Plotting is the graphical representation of data
- Plotting is an easier method to understand the nature of the data

### How to import matplotlib
- We mainly work with pyplot

1. plot():
- It is a method in mathplotlib
- It is used to plot the data in x and y axis

```python
import matplotlib.pyplot as plt
x = [1,2,3,45,5]
y = [10,20,30,40,50]

plt.plot(x,y)
plt.show()
```

# Approaches 
1. Functional approach
2. Object oriented approach 

1. Functional approach:
- It is begineer approach
- It is used to plot simple plots quickly

### Giving title and names to the axis
1. plt.title()
2. plt.xlabel('name')
3. plt.ylabel('name')


```python
plt.plot(x,y)
plt.title('Plot-1', fontdict={'font':'monospace', 'color':'red'})
plt.xlabel('Array X')
plt.ylabel('Array Y')

plt.show()
```

# Plotting multiple lines in a single plot


```python

plt.plot(x,y, color='red', ls='-', lw=3, marker='<', markerfacecolor='yellow', markersize=6,
         markeredgewidth=2, markeredgecolor='g', label='X vs Y')
plt.title('Plot-1', fontdict={'font':'monospace', 'color':'red'})
plt.xlabel('Array X')
plt.ylabel('Array Y')
plt.legend(loc=0, labelcolor='b', fontsize=10)

plt.show()
```

```python
x=[10,20,30,40,50,60]
y=[5,10,15,20,25,30]

plt.plot(x,y,color='red', ls='-', lw=3, marker='<', markerfacecolor='yellow', markersize=6,
         markeredgewidth=2, markeredgecolor='g', label='X vs Y')
plt.plot(y,x)
plt.plot(x*2, y*2)
plt.title('Plot-1', fontdict={'font':'monospace', 'color':'red'})
plt.xlabel('Array X')
plt.ylabel('Array Y')
plt.legend(loc=0, labelcolor='b', fontsize=10)

plt.show()
```

# Object oriented approach:

### Creating a figure
- plt.figure()

### Creating axes inside a figure
- fig.add_axes([left_boundry, bottom_boundry, wedth, height])

```python
import pandas as pd
ti_ds = pd.read_csv("Titanic.csv")
```

```python
f = plt.figure()
f.add_axes( [.1,.1,.5,.5] )
f.add_axes( [.7,.1,.5,.5] )
f.add_axes( [.7,.7,.8,.5] )
f.add_axes( [.1,.6,.5,.5] )
plt.show()
```

```python
f = plt.figure()
f.add_axes( [0.1,0.1,0.7,0.7] )
f.add_axes( [0.2,0.2,0.5,0.5] )
plt.show()
```
