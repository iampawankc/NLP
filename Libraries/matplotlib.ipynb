### How to import matplotlib

1. plot():

# Plotting multiple lines in a single plot


```python

plt.plot(x,y, color='red', ls='-', lw=3, marker='<', markerfacecolor='yellow', markersize=6,
         markeredgewidth=2, markeredgecolor='g', label='X vs Y')
plt.title('Plot-1', fontdict={'font':'monospace', 'color':'red'})
plt.xlabel('Array X')
plt.ylabel('Array Y')
plt.legend(loc=0, labelcolor='b', fontsize=10)

plt.show()
```

```python
x=[10,20,30,40,50,60]
y=[5,10,15,20,25,30]

plt.plot(x,y,color='red', ls='-', lw=3, marker='<', markerfacecolor='yellow', markersize=6,
         markeredgewidth=2, markeredgecolor='g', label='X vs Y')
plt.plot(y,x)
plt.plot(x*2, y*2)
plt.title('Plot-1', fontdict={'font':'monospace', 'color':'red'})
plt.xlabel('Array X')
plt.ylabel('Array Y')
plt.legend(loc=0, labelcolor='b', fontsize=10)

plt.show()
```

## Giving title and axes names in object oriented approach
- set_title('title')
- set_xlabel('axis_name')
- set_ylabel('axis_name')

```python
import numpy as np
from matplotlib import pyplot as plt

arr = np.arange(10)
arr2 = arr*2
```

```python
f = plt.figure(figsize=(10,10), dpi=100)
a = f.add_axes([0.1,0.1,0.5,0.5])
b = f.add_axes([0.7,0.1,0.5,0.5])

a.plot(arr, arr2, color='r', marker='o')
a.set_title('Plot 1')
a.set_xlabel('arr1')
a.set_xlabel('arr2')

b.plot(arr2, arr)
b.set_title('Plot 2')
b.set_xlabel('arr2')
b.set_xlabel('arr1')

plt.show()
```

# Boxplot
- It is used to plot statistical summary of the data (discrete column)
- boxplot works on IQR method 

Formula
- upperbound = Q3+1.5*IQR
- IQR= Q3-Q1
- lowerbound = Q1-1.5*IQR

```python
ti_ds.age.isna().sum()
```

```python
ti_ds.age.fillna(ti_ds.age.mean(), inplace=True)
```

```python
plt.boxplot(ti_ds.age)
plt.show()
```

```python
q3 = ti_ds.age.quantile(0.75)
print(q3)
q1 = ti_ds.age.quantile(0.25)
print(q1)
iqr = q3-q1
upper=q3+1.5*iqr
print(upper)
lower=q1-1.5*iqr
print(lower)
```

```python
#Removed the outliers
ti_ds.age.clip(lower, upper, inplace=True)
plt.boxplot(ti_ds.age)
plt.show()
```
